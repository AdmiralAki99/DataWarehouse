```python
# Establish database connection
def get_db_connection():
    return pyodbc.connect(
        "Driver={ODBC Driver 17 for SQL Server};"
        "Server=localhost;"
        "Database=CS689Project;"
        "Trusted_Connection=yes;"
    )

# Fetch existing active records from the database
def fetch_existing_hospital_data(conn):
    query = """
    SELECT ProviderCCN, HospitalName, City, StateCode, County, RuralVsUrban, EffectiveDate, ExpirationDate, Is_Current
    FROM HospitalDim WHERE Is_Current = 1
    """
    return pd.read_sql(query, conn)

# Check if a record has changes
def has_record_changed(existing_record, new_record):
    return (
        (existing_record['HospitalName'] != new_record['HospitalName']) or
        (existing_record['City'] != new_record['City']) or
        (existing_record['StateCode'] != new_record['StateCode']) or
        (existing_record['County'] != new_record['County']) or
        (existing_record['RuralVsUrban'] != new_record['RuralVersusUrban'])
    )

# Update existing record to expire it
def expire_record(cursor, provider_ccn):
    cursor.execute("""
        UPDATE HospitalDim
        SET ExpirationDate = ?, Is_Current = 0
        WHERE ProviderCCN = ? AND Is_Current = 1
    """, (datetime.now().strftime('%Y-%m-%d'), provider_ccn))
    print(f"Rows affected for ProviderCCN {provider_ccn}: {cursor.rowcount}")


# Insert a new record
def insert_new_record(cursor, row):
    cursor.execute("""
        INSERT INTO HospitalDim (
            ProviderCCN, HospitalName, City, StateCode, County, RuralVsUrban, 
            EffectiveDate, ExpirationDate, Is_Current
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, (
        row['ProviderCCN'], row['HospitalName'], row['City'], row['StateCode'], row['County'],
        row['RuralVersusUrban'], datetime.now().strftime('%Y-%m-%d'), '9999-12-31', 1
    ))

# Process incoming hospital data
def process_hospital_data(new_data, existing_data, conn):
    cursor = conn.cursor()
    for _, row in new_data.iterrows():
        # Check if the ProviderCCN exists in the current active data
        existing_record = existing_data[existing_data['ProviderCCN'] == row['ProviderCCN']]

        if not existing_record.empty:
            # Check if any attributes have changed
            if has_record_changed(existing_record.iloc[0], row):
                expire_record(cursor, row['ProviderCCN'])
                insert_new_record(cursor, row)
        else:
            # New record: Insert directly
            insert_new_record(cursor, row)
    conn.commit()
    cursor.close()

# Main function to orchestrate the SCD2 update
def main():
    # Example DataFrame to simulate delta data
    new_hospital_data = pd.DataFrame({
        'ProviderCCN': [10045, 102, 103],  # Existing and new ProviderCCNs
        'HospitalName': ['Hospital A', 'Hospital B Updated', 'Hospital C'],
        'City': ['Los Angeles', 'New York', 'Dallas'],
        'StateCode': ['CA', 'NY', 'TX'],
        'County': ['Los Angeles', 'Queens', 'Dallas'],
        'RuralVersusUrban': ['Urban', 'Urban', 'Rural']
    })

    # Establish database connection
    conn = get_db_connection()

    # Fetch existing data
    existing_hospital_data = fetch_existing_hospital_data(conn)

    # Process the data
    process_hospital_data(new_hospital_data, existing_hospital_data, conn)

    # Close connection
    conn.close()

# Execute the main function
if __name__ == "__main__":
    main()
```

```python
def load_staging_table(df, table_name, conn):
    """
    Load the DataFrame into the specified staging table.
    """
    cursor = conn.cursor()
    columns = ', '.join(df.columns)
    placeholders = ', '.join(['?'] * len(df.columns))
    sql = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"
    for _, row in df.iterrows():
        cursor.execute(sql, tuple(row))
    conn.commit()

def fetch_changed_records(conn):
    """
    Fetch records that have changes by comparing staging and main tables.
    """
    cursor = conn.cursor()
    # Example for detecting changes in HospitalDim:
    query = """
        SELECT s.*
        FROM Staging_HospitalInfo s
        LEFT JOIN HospitalDim h
        ON s.ProviderCCN = h.ProviderCCN
        WHERE h.HospitalName <> s.HospitalName
           OR h.City <> s.City
           OR h.StateCode <> s.StateCode
           OR h.County <> s.County
           OR h.RuralVsUrban <> s.RuralVsUrban
           OR h.ProviderCCN IS NULL; -- New records
    """
    cursor.execute(query)
    # Fetch and convert to a DataFrame for processing
    columns = [desc[0] for desc in cursor.description]
    records = cursor.fetchall()
    return pd.DataFrame.from_records(records, columns=columns)
```
